# Simple Logistic Regression

```python
final_df= df[df['class']!='Iris-setosa']
```

```python
final_df.head()
```

```python
#Outlier check
sns.pairplot(final_df,hue='class')
```

```python
sns.boxplot(final_df)
```

### As we can see that sepal_length_cm and sepal_width_cm has outliers

```python
final_df['sepal_length_cm'].hist()
```

```python
#It can be observed from the plot, that for 5 data points values are below 1 and they seem to be outliers. So, these data points are considered to be in 'm' and are converted to 'cm'.
final_df.loc[final_df.sepal_length_cm<1]=final_df["sepal_length_cm"]*100
```

```python
final_df["sepal_length_cm"].hist()
```

```python
final_df["sepal_width_cm"].hist()
```

```python
final_df.loc[final_df.sepal_width_cm>5]=final_df.drop(columns="sepal_width_cm")
```
